# **CALCULANDO A INTEGRAL DE Nm3/h EM RELAÇÃO A h, REFERENTE AO GRÁFICO.**

---



```python
# CALCULANDO A INTEGRAL DOS DADOS DE FLUXO DE MASSA EM RELAÇÃO ÀS HORAS DECIMAIS PELO MÉTODO DO TRAPÉZIO.
integral_numerica = np.trapz(dados_07dez['M_Fluxo_[Nm3/h]'], x=dados_07dez['Hora_decimal'])

print("Quantidade de Fluxo de Massa em Nm3/h: ", integral_numerica)

# MULTIPLICANDO O VALOS DA INTEGRAL POR 0,0899 PARA TRAZER A QUANTIDADE DE H2 EM KG.
qdd_H2 = integral_numerica * 0.0899

print("Quantidade de H2 produzido em kg: ", qdd_H2)

# COMPARANDO A RPODUÇÃO DE H2 CALCULADO E ENCONTRADO.
comparacao =  0.25 / qdd_H2

print("Relação de antes e depois H2/H2: ", comparacao)

# TRAZENDO O PERCENTUAL DE APROXIMAÇÃO ENTRE A RPODUÇÃO DE H2 CALCULADO E ENCONTRADO.
comparacao = comparacao * 100

# FORMATANDO O DADO DE COMPARAÇÃO
comparacao_format = "{:.3f}%".format(comparacao)

print("Relação Percentual de antes e depois H2/H2: ",comparacao_format)
```

# **CALCULANDO OS PARÂMETROS DE H2 CALCULADO E H2 ENCONTRADO E SUAS RAZÕES.**
---



```python
# DADOS DE H2 CALCULADO E SEUS PARÂMETROS.
qdd_h2_calculado = [0.25, 0.26, 0.26, 0.25, 0.26]

media_H2_cal = round(statistics.mean(qdd_h2_calculado), 4)
print("Média H2 Calculado:", media_H2_cal)
desvio_padrao_H2_cal = round(statistics.stdev(qdd_h2_calculado), 4)
print("Desvio padrão (amostral) H2 Calculado:", desvio_padrao_H2_cal)

# DADOS DE H2 ENCONTRADO E SEUS PARÂMETROS.
qdd_h2_encontrado = [0.327, 0.293, 0.329, 0.293, 0.313]

media_H2_enc = round(statistics.mean(qdd_h2_encontrado), 4)
print("Média H2 Encontrado:", media_H2_enc)
desvio_padrao_H2_enc = round(statistics.stdev(qdd_h2_encontrado), 4)
print("Desvio padrão (amostral) H2 Encontrado:", desvio_padrao_H2_enc)

# CRIANDO UMA LISTA PARA ARMAZENAR AS RAÇÕES DE H2.
proporcoes = []

# CALCULANDO A RAZÃO PARA CADA PAR DE VALORES CORRESPONDENTES.
for calc, encontrado in zip(qdd_h2_calculado, qdd_h2_encontrado):
    proporcao = encontrado / calc
    proporcoes.append(proporcao)

# CALCULANDO A MÉDIA DAS RAZÕES.
proporcao_media_H2 = sum(proporcoes) / len(proporcoes)
proporcao_media_H2 = round(proporcao_media_H2, 3)

# EXIBINDO AS RAZÕES.
print("Razões entre os grupos de dados:")
for i, proporcao in enumerate(proporcoes, 1):
    print(f"Razão {i}: {proporcao}")

print("Proporção média entre as razões:", proporcao_media_H2)

print("---------------------------------------------------------------------------")

# MULTIPLICANDO CADA VALOR POR 1.215.
qdd_h2_calculado_new = [x * 1.215 for x in qdd_h2_calculado]

media_new = round(statistics.mean(qdd_h2_calculado_new), 4)
desvio_padrao_new = round(statistics.stdev(qdd_h2_calculado_new), 4)

print("Média dos Novos Dados H2:", media_new)
print("Desvio Padrão dos Novos Dados H2:", desvio_padrao_new)
```
