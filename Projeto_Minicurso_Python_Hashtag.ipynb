# Desafio: importar uma base de dados do google drive, tratar seus dados e calcular o faturamento por loja, a quantidade de produtos por loja e o ticket médio, todos de forma tabular, com o maior valor em cima e o menor na parte de baixo.

```python
# Passo 01: Importar a base de dados

import os 
import pandas as pd
import openpyxl                 # O "openpyxl" faz a ligação entre o python e 
                                # arquivos do tipo excel. Aqui no google colab 
                                # não é necessário porque já vem esta biblioteca.

tabela = pd.read_excel("/content/drive/MyDrive/CÓDIGOS E BANCO DE DADOS PARA OS CURSOS PYTHON/Minicurso Python Hashtag/Vendas.xlsx")

# pd.set_option("display.max_columns", None) # O "pd.set_option(opção, valor)", 
# o "display.max_columns" significa mostrar máximo de colunas. Já o "None" é nenhum, 
# pois ele é o limitador, se desejamos mostrar todas as colunas, então não queremos
# um máximo, e sim todas elas.

display(tabela)

# O Erro "UnicodeDecodeError" significa um erro no qual existe 
# caracteres aos quais o python não conhece, como o "ç" no português.
```

>>> Na resposta acima se tem toda a tabela trazida do google drive.

Passo 02: Visualizar a base de dados

```python
tabela = tabela[["ID Loja", "Produto", "Quantidade", "Valor Unitário", "Valor Final"]]

display(tabela)
```

>>> Na resposta acima tem-se a tabela de forma organizada, trazendo apenas as colunas desejas.

Passo 03: Calcular o faturamento por loja

```python
tabela_fat = tabela.groupby("ID Loja").sum()
tabela_fat = tabela_fat[["Valor Final"]]
tabela_fat01 = tabela_fat
tabela_fat = tabela_fat[["Valor Final"]].sort_values(by="Valor Final", ascending=False)
display(tabela_fat)
```

>>> Na resposta acima temos a tabela de ID Loja e seu Valor Final, ou seja, uma tabela do Faturamento e a ordem descrescente dos valores de cima para baixo.

Passo 04: Calcular a quantidade de produtos vendidos por loja

```python
tabela_qtd = tabela.groupby("ID Loja").sum()
tabela_qtd = tabela_qtd[["Quantidade"]]
tabela_qtd01 = tabela_qtd
tabela_qtd = tabela_qtd[["Quantidade"]].sort_values(by="Quantidade", ascending=False)
display(tabela_qtd)
```

>>> Na resposta acima temos a tabela de ID Loja e sua Quantidade, ou seja, uma tabela da quantidade dos produtos vendidos e a ordem descrescente de cima para baixo.

Passo 05: Calcular o ticket médio por produto em cada loja

```python
ticket_medio = (tabela_fat01["Valor Final"] / tabela_qtd01["Quantidade"]).to_frame()
# O "().to_frame()" faz com que os dados que estão jogados se tornem
# dados em uma tabela, pois quando se faz uma operação de uma coluna com outra, 
# os dados são gerados, mas não tabelados.

print("_" * 50) # Gera uma string de caracteres que estão dentro dos aspas e 
                # multiplica pela quantidade desejada.
display(ticket_medio)
```

>>> Na resposta acima temos a tabela de ID Loja com seu valor de ticket médio em ordem descrescente dos valores de cima para baixo. Foi necessário a criação das variávies "tabela_fat01" e "tabela_qtd01" para manter a sequência dos dados na tabela, e assim foi possível a divisão correta dos dados de Faturamento pela sua Quantidade.

# Ordene os dados do ticket médio para melhor compreensão.

```python
ticket_medio_ordem = ticket_medio[[0]].sort_values(by= 0, ascending=False)

display(ticket_medio_ordem)
```

>>> Para finalizar, foi colocado em tabela o ticket médio de forma organizada, do maior valor no topo para o menor valor na parte de baixo.
