# O desafio é: Colete dados de um banco de dados do drive no formato csv, busque estes dados, importe-os, trate-os e compile, e calcule o produto mais vendido, o produto que mais faturou, a loja que mais vendeu, e por fim, faça um gráfico de colunas mostrando seus faturamentos.

# Passo 01 - Percorrer todos os arquivos da pasta base de dados (Pasta Vendas)

```python
# Dica: sempre colocar as importações de bibliotecas no começo para melhor organização.

import os               # Trabalha com arquivos e pastas do pc ou de site.
import pandas as pd     # Trabalha com dados e base de dados. O 'as pd' é um
                        # apelido para o 'pandas', ou seja, pd=pandas.

lista_arquivo = os.listdir("/content/drive/MyDrive/CÓDIGOS E BANCO DE DADOS PARA OS CURSOS PYTHON/Vendas") # O '.listdir' vai listar todo um diretório (pasta). 
                                                            # Dentro dos parênteses vai a localização da pasta, 
                                                            # entre parênteses. O 'lista_arquivo' é a variável.
print(lista_arquivo)
display(lista_arquivo)
```

---
Na resposta acima, o "print(lista_arquivo)" traz uma lista de arquivos de moto vetor com todos os nomes dos arquivos de uma pasta, e difere do "display(lista_arquivo)" pelo fato que neste último ele traz uma lista de arquivos mais organizado, de forma mais aprensentavel. É isso que o "display" faz.

---

# Passo 02 - Importar a base de dados de vendas

```python
tabela_total = pd.DataFrame()   # O '.DataFrame' é nada menos que uma tabela do python.
                                # E  criamos uma tabela vazia.

for arquivo in lista_arquivo:   # Sempre assim: for variável in lista_de_itens:
    if "Vendas" in arquivo:     # Sempre assim: if condição: 
                                # Se tem Vendas no nome do arquivo, eu quero.
        print(f"/content/drive/MyDrive/CÓDIGOS E BANCO DE DADOS PARA OS CURSOS PYTHON/Vendas/{arquivo}")
        # ou também: 
        # print("/content/drive/MyDrive/Vendas/{}".format(arquivo))
    # ou : 
    # if not "Devolucoes" in arquivo:
        # print(f"/content/drive/MyDrive/Vendas/{arquivo}")
        tabela = pd.read_csv(f"/content/drive/MyDrive/CÓDIGOS E BANCO DE DADOS PARA OS CURSOS PYTHON/Vendas/{arquivo}")
        tabela_total = tabela_total.append(tabela)      # O .append é adicionar. 
                                                        # x = x.append(y) , isto é, adiciona y em x
display(tabela_total)
```

---
Na resposta acima, o "print(f"/content/drive/MyDrive/Vendas/{arquivo}")" traz o nome do caminho de cada arquivo dentro da pasta desejada.

---
O "display(tabela)" traz como resposta a tabela de dados completa, com todos os arquivos de dados dentro de uma só tabela, e isso foi graças ao comando "tabela_total = tabela_total.append(tabela)".

---



# Passo 03 - Calcular o produto mais vendido (em quantidade)

```python
tabela_produtos = tabela_total.groupby("Produto").sum() # O '.groupby' agrupa uma coluna escolhida, e exige que você 
                                                        # decida o que fazer com as outras colunas. Neste caso, 
                                                        # o '.sum' faz com que somamos os valores.

display(tabela_produtos)  # Vai retornar todas as colunas da tabela.
tabela_produtos = tabela_produtos[["Quantidade Vendida", "Preco Unitario"]] # Vai retonrar as colunas escolhidas.
                                                                            # E obrigatoriamente, por ser mais de 
                                                                            # uma coluna, tem que ter dois colchetes.
tabela_produtos = tabela_produtos[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending=False)
# tabela_produtos = tabela_produtos[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending=True)
# o 'sort_values' ordena os dados, e ascending=True traz em ordem crescente e ascending=False traz em ordem descrescente.
display(tabela_produtos)
tabela_produtos = tabela_produtos["Quantidade Vendida"]

display(tabela_produtos)        # Quando se tem apenas 1 coluna, pode-se passar 1 colchete, 
                                # porém vai trazer dados mais "feios".
```

---
O Na resposta acima, o primeiro "display(tabela_produtos)" retorna toda a tabela com os dados calculados, neste caso, somados por causa do comando ".sum()".

---
O segundo "display(tabela_produtos)" nos traz uma tabela com apenas uma coluna desejada e organizada de forma crescente os dados.

---
No terceiro "display(tabela_produtos)" a tabela vem igual a segunda, porém com dados não em tabela (modo visual). Foi feito apenas para mostrar a diferença entre o segundo e o terceiro "display(tabela_produtos)" quando se coloca um ou dois [].

---

# Passo 04 - Calcular o produto que mais faturou (em faturamento)

```python
# Uma nova coluna tem que ser feita!
tabela_total["Faturamento"] = tabela_total["Quantidade Vendida"]*tabela_total["Preco Unitario"]
# Aqui multiplicamos cada valor da coluna Quantidade Vendida com cada valor da coluna Preco Unitario, surgindo uma
# nova coluna de Faturamento.

tabela_faturamento = tabela_total.groupby("Produto").sum()

tabela_faturamento = tabela_faturamento[["Faturamento"]].sort_values(by="Faturamento", ascending=False)

display(tabela_faturamento)
```

---
Podemos perceber que no resultado acima temos uma tabela com o produtos faturados, e de forma crescente os dados de baixo pra cima, no topo se encontra o item iPhone, que foi o de maior faturamento.

---

# Passo 05 - Calcular a loja/cidade que mais vendeu (em faturamento)

```python
 tabela_loja = tabela_total.groupby("Loja").sum()

tabela_loja = tabela_loja[["Faturamento"]].sort_values(by="Faturamento", ascending=False)

display(tabela_loja)
```

---
No resultado acima temos as lojas que mais vendeu, e de forma crescente de baixo pra cima, e no topo se encontra a loja de São Paulo como a que mais vendeu.

# Criar um gráfico/dashboard

```python
import plotly.express as px     # O o 'plotly.express' trabalha com gráficos. 
                                # O 'as px é um apelido para o 'plotly.express', 
                                # ou seja, px=plotly.express.

grafico = px.bar(tabela_loja, x = tabela_loja.index , y = "Faturamento")    # Aqui cria o gráfico. 

# Quando se faz 'tabela_loja = tabela_total.groupby("Loja").sum()', a coluna Loja dentro de 'groupby', ela se torna
# o índice da tabela, e não o título, por isso, quando for mostrar o gráfico, a coordenada tem que ser 
# x = nome_da_tabela.index, para vir seu índice.

grafico.show() # Aqui ele mostra o gráfico.
```

---
No gráfico acima temos o faturamento de cada loja de forma visul e descrescente, começando por São Paulo a loja que mais vendeu para a que menos vendeu, sendo esta a de Belo Horizonte.
