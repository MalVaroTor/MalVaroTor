# **CALCULANDO AS POTÊNCIAS A E B E ADICIONANDO AO DB.**

---



```python
# TRAZENDO O ÍNDICE DA COLUNA 'Tensao_A_[V]'.
indice_tensao_A = db_ps_estudo.columns.get_loc('Tensao_A_[V]')

# INSERINDO A COLUNA 'Potencia_A_[kW]' DEPOIS DA COLUNA 'Tensao_A_[V]'.
db_ps_estudo.insert(indice_tensao_A + 1, 'Potencia_A_[kW]', db_ps_estudo['Corrente_A_[A]'] * db_ps_estudo['Tensao_A_[V]'] / 1000)

# TRAZENDO O ÍNDICE DA COLUNA 'Tensao_B_[V]'.
indice_tensao_B = db_ps_estudo.columns.get_loc('Tensao_B_[V]')

# INSERINDO A COLUNA 'Potencia_B_[kW]' DEPOIS DA COLUNA 'Tensao_B_[V]'.
db_ps_estudo.insert(indice_tensao_B + 1, 'Potencia_B_[kW]', db_ps_estudo['Corrente_B_[A]'] * db_ps_estudo['Tensao_B_[V]'] / 1000)

display(db_ps_estudo)
```

# **GERANDO GRÁFICO DE POTÊNCIAS A E B.**

```python
# CRIANDO A FIGURA.
plt.figure(figsize=(10, 6))

# PLOTANDO OS DADOS DO GRÁFICO.
plt.plot(db_ps_estudo['Data/Hora'], db_ps_estudo['Potencia_A_[kW]'], label='Potência A')
plt.plot(db_ps_estudo['Data/Hora'], db_ps_estudo['Potencia_B_[kW]'], label='Potência B')

# ADICIONANDO OS EIXOS.
plt.xlabel('Horas')
plt.ylabel('Potências [kW]')

# ADICIONANDO O TÍTULO.
plt.title(f'Potências A e B em [kW] ao longo do tempo no dia {var_data_estudo}')

# # ADICIONANDO A LEGENDA.
plt.legend()

# FORMATANDO O EIXO X.
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%H:%M:%S'))

# INCLINANDO O EIXO X.
# plt.xticks(rotation=45)

# EXIBINDO O GRÁFICO.
plt.tight_layout()
plt.show()
```

# **CRIANDO A COLUNA "Hora_Decimal."**

---



```python
# CRIANDO A COLUNA 'Hora_decimal'.
db_ps_estudo['Hora_decimal'] = db_ps_estudo['Data/Hora'].dt.hour + db_ps_estudo['Data/Hora'].dt.minute / 60 + db_ps_estudo['Data/Hora'].dt.second / 3600

display(db_ps_estudo)
```

# **CALCULANDO A INTEGRAL DO GRÁFICO DE POTÊNCIA EM RELAÇÃO AO TEMPO PARA DETERMINAR A ENERGIA.**

---



```python
energia_a = np.trapz(db_ps_estudo['Potencia_A_[kW]'], x=db_ps_estudo['Hora_decimal'])
energia_a = round(energia_a, 3)

energia_b = np.trapz(db_ps_estudo['Potencia_B_[kW]'], x=db_ps_estudo['Hora_decimal'])
energia_b = round(energia_b, 3)

print(f"Quantidade de Energia do Eletrolisador A em [kWh] no dia {var_data_estudo}: ", energia_a)

print(f"Quantidade de Energia do Eletrolisador B em [kWh] no dia {var_data_estudo}: ", energia_b)
```

# **LIMITANDO AGORA ALÉM DA DATA ESCOLHIDA, O TEMPO DE OPERAÇÃO TAMBÉM.**

---



```python
# Especificar o intervalo de horário
hora_inicio = input("Digite a hora inicial do PowerSet de produção no formato 'HH:MM:SS': ")
hora_fim = input("Digite a hora final do PowerSet de produção no formato 'HH:MM:SS': ")

# Filtrar os dados dentro do intervalo de horário
db_hr_estudo = db_ps_estudo[(db_ps_estudo['Data/Hora'].dt.time >= pd.to_datetime(hora_inicio).time()) &
                                                (db_ps_estudo['Data/Hora'].dt.time <= pd.to_datetime(hora_fim).time())]

display(db_hr_estudo)
```
